# Implement Preprocessing Data

## Symbol puntuation removal

```python
import re

# Fungsi untuk menghapus simbol dan tanda baca
def remove_symbols(text):
    # Ubah ke string (jaga-jaga)
    text = str(text)
    # Hapus karakter \n, \t, dll
    text = re.sub(r'\s+', ' ', text)
    # Hapus semua simbol dan tanda baca kecuali huruf dan angka
    text = re.sub(r'[^\w\s]', '', text)
    return text

# Terapkan ke kolom review_text
df['symbol_puntuation_removal'] = df['review'].apply(remove_symbols)

# Lihat hasil
print(df[['review', 'symbol_puntuation_removal']].head())
```

## Case Folding

```python
# Ubah semua teks di kolom cleaned_review menjadi lowercase
df['case_folding'] = df['symbol_puntuation_removal'].str.lower()

# Cek hasilnya
print(df[['symbol_puntuation_removal', 'case_folding']].head())
```

## Tokenized

```python
# Tokenisasi manual: pisah kata berdasarkan spasi
df['tokenized_review'] = df['case_folding'].apply(lambda x: x.split())

# Cek hasilnya
print(df[['case_folding', 'tokenized_review']].head())
```

## Normalization

```python
from collections import Counter

# Load kamus kata baku dari file
with open("kata-dasar-all.txt", "r", encoding="utf-8") as f:
    kata_baku_set = set(line.strip() for line in f)

# Ambil semua token dari kolom tokenized_review
all_tokens = [token for tokens in df['tokenized_review'] for token in tokens]

# Deteksi kata tidak baku
kata_tidak_baku = [token for token in all_tokens if token not in kata_baku_set]

# Hitung frekuensi
frekuensi_tidak_baku = Counter(kata_tidak_baku)

# Tampilkan Top 30 kata tidak baku
print("Top 30 kata tidak baku:")
for kata, jumlah in frekuensi_tidak_baku.most_common(100):
    print(f"{kata}")
```

```python
# Baca kamus normalisasi
df_kamus = pd.read_excel('kamus_normalisasi.xlsx')

# Buat dictionary hanya untuk kata yang memiliki 'Kata Baku'
kamus_normalisasi = df_kamus.dropna(subset=["Kata Baku"]).set_index("Kata")["Kata Baku"].to_dict()

# Buat set kata-kata yang ingin dihapus (karena 'Kata Baku'-nya kosong)
kata_dihapus = set(df_kamus[df_kamus["Kata Baku"].isnull()]["Kata"].tolist())

# Fungsi normalisasi
def normalisasi_kata(token_list, kamus, hapus_kata):
    return [kamus.get(kata, kata) for kata in token_list if kata not in hapus_kata]

# Terapkan ke data
df['normalized_review'] = df['tokenized_review'].apply(lambda x: normalisasi_kata(x, kamus_normalisasi, kata_dihapus))
df[['tokenized_review', 'normalized_review']].head()
```

## Stopword

```python
from nltk.corpus import stopwords
```

```python
stop_words = set(stopwords.words('indonesian'))
def hapus_stopwords(token_list, stopword_list):
    return [kata for kata in token_list if kata not in stopword_list]
df['stopwords_removed'] = df['normalized_review'].apply(lambda x: hapus_stopwords(x, stop_words))
df[['normalized_review', 'stopwords_removed']].head()
```

## Stemming

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

```python
from tqdm import tqdm
tqdm.pandas()

stem_cache = {}

def stem_teks(teks):
    if isinstance(teks, list):
        teks = ' '.join(teks)
    
    stemmed = []
    for word in teks.split():
        if word in stem_cache:
            stemmed.append(stem_cache[word])
        else:
            s = stemmer.stem(word)
            stem_cache[word] = s
            stemmed.append(s)
    
    return ' '.join(stemmed)

df['stemmed_review'] = df['stopwords_removed'].progress_apply(stem_teks)
df[['stopwords_removed', 'stemmed_review']].head()
```

# Implementation Feature Extraction

## BoW

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Ambil korpus dari kolom 'stemmed_review'
corpus = df['stemmed_review'].tolist()

# Inisialisasi dan ekstraksi BoW
vectorizer = CountVectorizer(
    min_df=3,
    max_df=0.95,
    max_features=1500,
)
bow_matrix = vectorizer.fit_transform(corpus)

# Konversi ke array dan simpan ke DataFrame
bow_array = bow_matrix.toarray()
bow_df = pd.DataFrame(bow_array, columns=vectorizer.get_feature_names_out())

# Buat nama review seperti R_bow_1, R_bow_2, dst
bow_df.insert(0, 'Review', [f'R_bow_{i+1}' for i in range(len(bow_df))])

# Tambahkan kolom 'Vector' yang berisi daftar angka BoW untuk tiap dokumen
bow_df['Vector'] = bow_array.tolist()

# Tampilkan sebagian hasil
print(bow_df[['Review', 'Vector']].head())

# Lihat seluruh tabel: Review + fitur + Vector
print(bow_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Hitung total kemunculan setiap kata
word_freq = bow_array.sum(axis=0)
word_freq_series = pd.Series(word_freq, index=vectorizer.get_feature_names_out())

# Plot 20 kata teratas
plt.figure(figsize=(12, 6))
word_freq_series.sort_values(ascending=False).head(20).plot(kind='bar', color='orange')
plt.title("20 Kata Paling Sering dalam BoW")
plt.ylabel("Jumlah Kemunculan")
plt.xlabel("Kata")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Hasil ekstraksi fitur menggunakan metode **Bag of Words (BoW)** dinyatakan **valid** dan layak digunakan untuk pemodelan machine learning, berdasarkan beberapa indikator berikut:

| **Kriteria**                        | **Status** | **Bukti**                                                                                   |

🔍 Dengan ketiga indikator tersebut, dapat disimpulkan bahwa hasil BoW yang digunakan telah mencerminkan **representasi teks yang informatif dan tidak bias**, serta **relevan secara semantik** terhadap topik ulasan pelanggan hotel.

# Implementation Individual Algorithm + BoW

### Random Forest + BoW

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import numpy as np

aspek_sentimen = ['ac', 'air_panas', 'bau', 'general', 'kebersihan', 
                  'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- Random Forest for Aspect: {aspek} -----")

    # Ambil label untuk aspek ke-i
    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    # Inisialisasi dan latih model
    rf = RandomForestClassifier(
        n_estimators=100,
        max_depth=None,
        min_samples_leaf=2,
        class_weight='balanced',
        random_state=42
    )
    rf.fit(X_train, y_train_aspek)

    # Prediksi dan evaluasi
    y_pred = rf.predict(X_test)
    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
import numpy as np

aspek_sentimen = ['ac', 'air_panas', 'bau', 'general', 'kebersihan', 
                  'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- SVM for Aspect: {aspek} -----")

    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    svm = LinearSVC(class_weight='balanced', random_state=42)
    svm.fit(X_train, y_train_aspek)
    y_pred = svm.predict(X_test)

    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))
```

### Multinomial Naive Bayes + BoW

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
from imblearn.over_sampling import RandomOverSampler
import numpy as np

# Daftar aspek sentimen
aspek_sentimen = ['ac', 'air_panas', 'bau', 'general', 'kebersihan', 
                  'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

# Iterasi untuk masing-masing aspek sentimen
for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- Multinomial Naive Bayes for Aspect: {aspek} -----")

    # Ambil label untuk aspek ke-i (multi-label: satu classifier per aspek)
    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    # Terapkan teknik oversampling untuk menangani data tidak seimbang
    # Oversampling hanya dilakukan di data pelatihan
    ros = RandomOverSampler(random_state=42)
    X_train_resampled, y_train_resampled = ros.fit_resample(X_train, y_train_aspek)

    # Inisialisasi model Multinomial Naive Bayes
    mnb = MultinomialNB()
    
    # Latih model dengan data yang sudah di-resample
    mnb.fit(X_train_resampled, y_train_resampled)

    # Prediksi pada data uji
    y_pred = mnb.predict(X_test)

    # Evaluasi hasil klasifikasi
    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))
```

### Hybrid with imbalanced data solutions + BoW

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import ComplementNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
import numpy as np

aspek_sentimen = ['ac', 'air_panas', 'bau', 'general', 'kebersihan', 
                  'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

for i, aspek in enumerate(aspek_sentimen):
    print(f"\n===== Hybrid Classifier for Aspect: {aspek} =====")

    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    # Split ulang training menjadi train + validation khusus untuk stacking
    X_train_stack, X_val_stack, y_train_stack, y_val_stack = train_test_split(
        X_train, y_train_aspek, test_size=0.3, random_state=42
    )

    # Terapkan oversampling pada data train untuk menangani ketidakseimbangan
    ros = RandomOverSampler(random_state=42)
    X_train_stack_resampled, y_train_stack_resampled = ros.fit_resample(X_train_stack, y_train_stack)

    # Base models dengan penyesuaian
    base_models = [
        RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42),
        SVC(kernel='linear', probability=True, class_weight='balanced', random_state=42),
        MultinomialNB(),
        KNeighborsClassifier(n_neighbors=5)  # KNN tidak mendukung class_weight
    ]

    meta_features = []
    trained_models = []

    for model in base_models:
        # Bootstrapping
        X_boot, y_boot = resample(X_train_stack, y_train_stack, replace=True, random_state=42)
        model.fit(X_boot, y_boot)
        trained_models.append(model)

        # Prediksi validation set (gunakan proba kalau tersedia)
        if hasattr(model, "predict_proba"):
            preds = model.predict_proba(X_val_stack)
            # Ambil probabilitas untuk kelas positif jika binary
            if preds.shape[1] == 2:
                preds = preds[:, 1].reshape(-1, 1)
        else:
            preds = model.decision_function(X_val_stack)
            preds = preds.reshape(-1, 1)

        meta_features.append(preds)

    # Dataset meta untuk training meta-model
    X_meta_train = np.hstack(meta_features)
    y_meta_train = y_val_stack

    # Meta model
    meta_model = LogisticRegression(max_iter=1000, random_state=42)
    meta_model.fit(X_meta_train, y_meta_train)

    # Prediksi test set dengan trained base models
    meta_test_features = []
    for model in trained_models:
        if hasattr(model, "predict_proba"):
            preds = model.predict_proba(X_test)
            if preds.shape[1] == 2:
                preds = preds[:, 1].reshape(-1, 1)
        else:
            preds = model.decision_function(X_test)
            preds = preds.reshape(-1, 1)

        meta_test_features.append(preds)

    X_meta_test = np.hstack(meta_test_features)
    y_pred = meta_model.predict(X_meta_test)

    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))
```

### Comparison combination hybrid classifier + BoW

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import ComplementNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
import numpy as np

aspek_sentimen = ['ac', 'air_panas', 'bau', 'general', 'kebersihan', 
                  'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

# ===============================
# 1. STACKING CLASSIFIER
# ===============================
print("=== STACKING HYBRID CLASSIFIER ===")
for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- Stacking for Aspect: {aspek} -----")

    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    X_train_stack, X_val_stack, y_train_stack, y_val_stack = train_test_split(
        X_train, y_train_aspek, test_size=0.3, random_state=42
    )

   # Terapkan oversampling pada data train untuk menangani ketidakseimbangan
    ros = RandomOverSampler(random_state=42)
    X_train_stack_resampled, y_train_stack_resampled = ros.fit_resample(X_train_stack, y_train_stack)

    # Base models dengan penyesuaian
    base_models = [
        RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42),
        SVC(kernel='linear', probability=True, class_weight='balanced', random_state=42),
        MultinomialNB(),
        KNeighborsClassifier(n_neighbors=5)  # KNN tidak mendukung class_weight
    ]

    meta_features = []
    trained_models = []

    for model in base_models:
        X_boot, y_boot = resample(X_train_stack, y_train_stack, replace=True, random_state=42)
        model.fit(X_boot, y_boot)
        trained_models.append(model)

        if hasattr(model, "predict_proba"):
            preds = model.predict_proba(X_val_stack)
            preds = preds[:, 1].reshape(-1, 1) if preds.shape[1] == 2 else preds
        else:
            preds = model.decision_function(X_val_stack).reshape(-1, 1)

        meta_features.append(preds)

    X_meta_train = np.hstack(meta_features)
    y_meta_train = y_val_stack

    meta_model = LogisticRegression(max_iter=1000, random_state=42)
    meta_model.fit(X_meta_train, y_meta_train)

    meta_test_features = []
    for model in trained_models:
        if hasattr(model, "predict_proba"):
            preds = model.predict_proba(X_test)
            preds = preds[:, 1].reshape(-1, 1) if preds.shape[1] == 2 else preds
        else:
            preds = model.decision_function(X_test).reshape(-1, 1)

        meta_test_features.append(preds)

    X_meta_test = np.hstack(meta_test_features)
    y_pred = meta_model.predict(X_meta_test)

    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))


# ===============================
# 2. VOTING CLASSIFIER (Soft Voting)
# ===============================
print("\n=== VOTING HYBRID CLASSIFIER (Soft Voting) ===")
voting_clf = VotingClassifier(
    estimators=[
        ('rf', RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)),
        ('svc', SVC(kernel='linear', probability=True, class_weight='balanced', random_state=42)),
        ('nb', MultinomialNB()),
        ('knn', KNeighborsClassifier(n_neighbors=5))
    ],
    voting='soft'
)

for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- Voting for Aspect: {aspek} -----")
    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    voting_clf.fit(X_train, y_train_aspek)
    y_pred = voting_clf.predict(X_test)

    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))


# ===============================
# 3. WEIGHTED VOTING CLASSIFIER
# ===============================
print("\n=== WEIGHTED VOTING HYBRID CLASSIFIER ===")
weighted_voting_clf = VotingClassifier(
    estimators=[
        ('rf', RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)),
        ('svc', SVC(kernel='linear', probability=True, class_weight='balanced', random_state=42)),
        ('nb', MultinomialNB()),
        ('knn', KNeighborsClassifier(n_neighbors=5))
    ],
    voting='soft',
    weights=[3, 3, 1, 1]  # Give more weight to RF and SVC
)

for i, aspek in enumerate(aspek_sentimen):
    print(f"\n----- Weighted Voting for Aspect: {aspek} -----")
    y_train_aspek = y_train[:, i]
    y_test_aspek = y_test[:, i]

    weighted_voting_clf.fit(X_train, y_train_aspek)
    y_pred = weighted_voting_clf.predict(X_test)

    print(classification_report(y_test_aspek, y_pred, digits=3, zero_division=0))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data akurasi untuk setiap aspek dari 3 jenis hybrid classifier (Stacking, Voting, Weighted Voting)
aspek = ['ac', 'air_panas', 'bau', 'general', 'kebersihan',
         'linen', 'service', 'sunrise_meal', 'tv', 'wifi']

stacking_scores = [0.961, 0.967, 0.937, 0.906, 0.816,
                   0.882, 0.904, 0.958, 0.980, 0.982]

voting_scores = [0.963, 0.965, 0.934, 0.897, 0.829,
                 0.873, 0.867, 0.947, 0.969, 0.972]

weighted_scores = [0.963, 0.967, 0.939, 0.893, 0.827,
                   0.882, 0.871, 0.954, 0.982, 0.978]

# Buat DataFrame
df_compare = pd.DataFrame({
    'Aspect': aspek,
    'Stacking': stacking_scores,
    'Voting': voting_scores,
    'Weighted Voting': weighted_scores
})

# Plotting
plt.figure(figsize=(14, 6))
df_compare.set_index('Aspect').plot(kind='bar', figsize=(14, 6), colormap='Set2')
plt.title('Accuracy Comparison of Hybrid Classifiers per Aspect')
plt.ylabel('Accuracy')
plt.ylim(0.75, 1.0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Hitung rata-rata akurasi untuk setiap metode hybrid
avg_stacking = df_compare['Stacking'].mean()
avg_voting = df_compare['Voting'].mean()
avg_weighted = df_compare['Weighted Voting'].mean()

# Buat DataFrame rata-rata
df_avg = pd.DataFrame({
    'Hybrid Method': ['Stacking', 'Voting', 'Weighted Voting'],
    'Average Accuracy': [avg_stacking, avg_voting, avg_weighted]
})

# Visualisasi rata-rata
plt.figure(figsize=(8, 5))
plt.bar(df_avg['Hybrid Method'], df_avg['Average Accuracy'], color=['#66c2a5', '#fc8d62', '#8da0cb'])
plt.title('Average Accuracy of Hybrid Classifiers')
plt.ylabel('Average Accuracy')
plt.ylim(0.85, 0.98)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Tampilkan nilai rata-rata dalam bentuk tabel juga
df_avg
```
